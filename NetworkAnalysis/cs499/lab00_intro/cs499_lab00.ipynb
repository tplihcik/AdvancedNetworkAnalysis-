````python
import networkx as nx
````

Now that we have imported NetworkX, we can use it anywhere in our notebook. Let's start by creating a simple network using NetworkX's Graph class.  

````python
# Creating a new complex network with Graph
G = nx.Graph()

# Next, we will add a node with an ID of 1
G.add_node(1)

# We can also add multiple nodes at a time using a list
G.add_nodes_from([2,3])
````

Let's draw our growing network, _G_, as we start adding components to it. We need to use matplotlib.pyplot for this. You can install matplotlib using the following command in your terminal:

    pip3 install matplotlib

````python
import matplotlib.pyplot as plt

# Create a subplot in which to draw your network
plt.subplot(121)

# Draw G
nx.draw(G, with_labels=True, font_weight='bold')

# Show the drawing
plt.show()
````

Let's add some edges. There are a number of ways to add edges. We will explore adding unweighted and weighted edges as well as adding colors to edges.

````python
# Add edges
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(2,3)

# Create a subplot in which to draw your network
plt.subplot(122)

# Draw G
nx.draw(G, with_labels=True, font_weight='bold')

# Show the drawing
plt.show()
````

Now let's look at how to access nodes and edges from a Graph object. 

````python
nodes = G.nodes()
edges = G.edges()

print("These are the nodes:")
print(nodes)

print("These are the edges:")
print(edges)
````

Note that this is an undirected graph. We can create an directed graph in NetworkX as well. 

````python
# Create a directed graph
DG = nx.DiGraph()

# Add some weighted edges to the graph.
# The edge format is a triple with the endpoints and the weight.
# Note that we can use strings as nodes
DG.add_weighted_edges_from([("A", "B", 0.5), ("C", "A", 0.75)])

# Create a subplot in which to draw your network
plt.subplot(122)

# Draw G
nx.draw(DG, with_labels=True, font_weight='bold')

# Show the drawing
plt.show()
````

### Characterizing networks

An important way we will be using NetworkX is to characterize networks. Start generating an instance of the well-known Karate Club graph. 

````python
# There are some pre-defined, real-world networks
# that are accessible via the NetworkX API
G = nx.karate_club_graph()

# Use a spring layout to spread nodes apart for a drawing
pos = nx.spring_layout(G)

plt.figure(figsize=(12,8))
nx.draw_networkx(G, pos=pos, node_color="gray", node_size=640)
limits = plt.axis('off')
plt.show()
````
