## Part 1: Calculating the Shortest Path

Use Dijkstra's shortest path algorithm to calculate the shortest path between any two nodes in a graph. The function signature (function name and parameters) are provided. You can use the Karate Club graph as the graph input. 

### Check List for shortestPath() ###

- [ ] Calculates the length of the shortest path between the src and dst nodes.
- [ ] Does NOT use any of the NetworkX shortest path functions such as networkx.single_source_shortest_path_length()
- [ ] Is able to handle directed and undirected networks as input
- [ ] Will be tested against both directed and undirected networks

*Note that you do not need to support weighted networks, altough implementing this would make for an excellent stretch goal for this lab*

```python
#Collaborated with Adam Schilperoort
import networkx as nx



# If you would like to use the Karate Club graph to test with
G = nx.karate_club_graph()

#testing for directed graph 
#G = nx.DiGraph()



# shortestPath takes in
# G: A networkx graph (directed or undirected)
# src: The id of the src node
# dst: The id of the dst node
def shortestPath(G, src, dst):
    # TODO: Implement Dijkstra's shortest path algorithm
    
   
    
    
    #infinty value to init all distances too
    infinity = 9999999
    
    #list for visited nodes
    visitedNodes = []
    
    #list for unvisited nodes 
    unvisitedNodes = []
    
    #list of tuples: node, distance
    nodeDistanceList = []
    
    #iterator var for looping
    iterator = 0 
    

    
    #set the tuple value pairs for each node and its distance (always starting at infinity)
    for node in G.nodes:
        unvisitedNodes.append( node )
        
        #distance from the source node is always 0 or NULL
        if node == src:
            nodeDistanceList.append( ( node , 0 ) )
            
        else:
            nodeDistanceList.append( ( node , infinity ) )
        
        #display the tuple list
        #print( nodeDistanceList[ iterator ] )
        iterator += 1
        
    
        
    #loop through the unvisited nodes list until they have all been visited 
    while len(unvisitedNodes) != 0:

        #get node with shortest distance
        shortestNode = src 
        shortestDistance = infinity 

        for node in unvisitedNodes:

            if nodeDistanceList[ node ][ 1 ] < shortestDistance:
                shortestDistance = nodeDistanceList[ node ][ 1 ]
                shortestNode = node 

        #go to the shortest distance node
        currentNode = shortestNode

        

        #move current node from unvisited to visited
        visitedNodes.append( currentNode )
        unvisitedNodes.remove( currentNode )

        #get the current nodes neighbors
        currentNodeNeighbors = list( G.neighbors( currentNode ) )

       

        #iterate through the list of neighbors for the current node and find their distances
        for neighborNode in currentNodeNeighbors:
            
            #set distance from src node to current node 
            currentNodeDistance = nodeDistanceList[ neighborNode ][ 1 ]
            
            #set distance from current node to new node 
            newNodeDistance = nodeDistanceList[ currentNode ][ 1 ] + 1

            #if the distance to the new node is less than the distance currently listed
            if newNodeDistance < currentNodeDistance:

                #replace the old distance with the up to date shorter one 
                nodeDistanceList.remove( ( neighborNode , currentNodeDistance ) )
                nodeDistanceList.insert( neighborNode , ( neighborNode , newNodeDistance ) ) 


    #get result from tuple 
    result = nodeDistanceList[ dst ][ 1 ]
    
    #print the result 
    print( 'The result of the algorithm, from source ' + str( src ) + ' to ' + str( dst ) + ' is: '  + str( result ) )

    
    return result                
            
        
        
    
    
    
       
    
    
    
```

## Part 2: Calculate the Average Path Length

In this part of the lab, you will write a program to calculate the average path length associated with a network. You may use any of the functions you have implemented as part of this lab. 

- [ ] Calculates the average shortest path length for all pairs of nodes in the network. Returns a float representing this value.




```python
#Collaborated with Adam Schilperoort

import networkx as nx
G = nx.karate_club_graph()



# averagePath takes in
# G: A networkx graph (directed or undirected)
def averagePath(G):
    # TODO Implement averagePath
    
    #list for all of the nodes in the graph 
    listOfNodes = []
    
    #counter to use in average calculation 
    counter = 0 
    
    #length value of the paths
    length = 0 
    
    #fill the list with the nodes in the given graph 
    for node in G.nodes:
        listOfNodes.append( node )
    
    #loop through and run the shortest path algorithm
    for sourceIndex in range( 0 , len( listOfNodes ) ):
        for destIndex in range( (sourceIndex + 1) , len( listOfNodes ) ):
        
            length += shortestPath( G , sourceIndex , destIndex )
            counter += 1
            
    #calculate the final average of all lengths 
    averageLength = length / counter
    
    #print result 
    print( "Average Path Length for this graph is: " + str( averageLength ) )
        
    
    
averagePath(G)
    
```

## Part 3: Calculate the Diameter

In this part of the lab, you will write a program to calculate the diameter of a network. You may use any of the functions you have implemented as part of this lab. 

```python
#Collaborated with Adam Schilperoort 

# networkDiameter takes in
# G: A networkx graph (directed or undirected)
def networkDiameter(G):
    # TODO Implement networkDiameter
    
    #diameter is the max shortest path
    
    #load all the path lengths into a list and run a max 
    #function on it 
    
    #list for all the nodes in the grapb 
    listOfNodes = []
    
    
    counter = 0 
        
    #temp value used for comparison of current vs past path length 
    storageValue = 0 
    
    
    #fill the list with the nodes from G 
    for node in G.nodes:
        listOfNodes.append( node )
    
    #loop through and run the shortest path algorithm
    for sourceIndex in range( 0 , len( listOfNodes ) ):
        for destIndex in range( (sourceIndex + 1) , len( listOfNodes ) ):
        
            pathLength = shortestPath( G , sourceIndex , destIndex )            
            
            #if the current iterations path length is greater than the 
            #previous iteration's, then swap the values 
            if pathLength > storageValue:
                storageValue = pathLength
      
    #print the result 
    #To make it easy to find:
    print( "=======================================" )
    print( "The diameter of this graph is: " +str(storageValue) )
    
    #return the value 
    return storageValue

networkDiameter(G)
                
    
    
    
    
    
    
        
```
